```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

mpl.rcParams['font.size'] = 16
x = np.linspace(0, 10, 100)
y = 3 * np.exp(-x) + np.random.normal(0, 0.07, size=x.shape)
plt.plot(x, y, label='$3 e^{-x}$ with noise', color='blue', lw=2)
y = np.sqrt(x)
plt.plot(x, y, label=r'$\sqrt{x}$', color='red', lw=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('A very important plot')
plt.legend()
plt.savefig('line.pdf', bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = 0.5 * x + np.random.normal(0, 0.5, size=x.shape)
plt.scatter(x, y, label='0.5x + noise', color='blue', marker='o')
y = 1.5 * x + np.random.normal(0, 0.5, size=x.shape)
plt.scatter(x, y, label='1.5x + noise', color='red', marker='^')
plt.xlabel('x')
plt.ylabel('y')
plt.title('An even more important plot')
plt.legend()
plt.savefig('scat.pdf', bbox_inches='tight')
plt.show()
```
